```python
import stpsf
import webbpsf

import matplotlib.pyplot as plt
import numpy as np
```

```python
wfi_stpsf = stpsf.roman.WFI()
wfi_webbpsf = stpsf.roman.WFI()
```

```python
collector_stpsf = []
collector_webbpsf = []

for ii in range(1, 18):
    det = "SCA{:02}".format(ii)
    wfi_stpsf.detector = det
    wfi_webbpsf.detector = det

    temp_stpsf = wfi_stpsf.calc_psf()
    temp_webbpsf = wfi_webbpsf.calc_psf()

    collector_stpsf.append(temp_stpsf)
    collector_webbpsf.append(temp_webbpsf)
```

```python
for ii in range(0, 17):
    fig, ax = plt.subplots(1, 3)
    
    stpsf_plot = collector_stpsf[ii][0].data
    webbpsf_plot = collector_webbpsf[ii][0].data

    ax[0].set_title("STPSF")
    ax[0].imshow(stpsf_plot)

    ax[1].set_title("WEBBPSF")
    ax[1].imshow(webbpsf_plot)

    ax[2].set_title("Diff: %i"%(np.any(stpsf_plot - webbpsf_plot)))
    ax[2].imshow(stpsf_plot - webbpsf_plot)
```

```python
for ii in range(0, 17):
    for jj in range(0, 17):
        if ii == jj:
            continue

        psf_ii = collector_stpsf[ii][0].data
        psf_jj = collector_stpsf[jj][0].data

        diff = psf_ii - psf_jj

        if not np.any(diff):
            print("Duplicate found")

        else:
            print("Non-duplicate")

        # plt.imshow(diff)
        # plt.title("Det diff %i"%(np.any(diff)))
        # plt.show()
```
